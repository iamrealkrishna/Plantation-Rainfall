```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"F:\misc\UPASI Project\final\\DailyMet.xlsx")
df.head(10)
```

```python
df['Rainfall Inches']=df['Rainfall Inches'].replace(['Nil'],0)
df['Rainfall Inches']=df['Rainfall Inches'].replace(['traces'],0.01)
```

```python
df.head()
```

```python
df['Rainfall Inches']=df['Rainfall Inches'].replace(['Traces'],0.01)
df['Rainfall Inches']=df['Rainfall Inches'].replace(['mild 0'],0.0001)
df.head(10)
```

```python
df['Rainfall Inches']=df['Rainfall Inches'].replace(['Traces'],0.01)
df['Rainfall Inches']=df['Rainfall Inches'].replace(['mild 0'],0.0001)
df['Rainfall Inches']=df['Rainfall Inches'].replace(['Frost'],0)
df['Rainfall Inches']=df['Rainfall Inches'].replace(['frost'],0)
df['Rainfall Inches']=df['Rainfall Inches'].replace(['mild frost'],0)
df.head(10)
```

```python
df.drop(df[(df['Day']>28) & (df['Month']==2) & (df['Year']%4!=0)].index,inplace=True)
df.drop(df[(df['Day']>29) & (df['Month']==2) & (df['Year']%4==0)].index,inplace=True)
df.drop(df[(df['Day']>30) & ((df['Month']==4)|(df['Month']==6)|(df['Month']==9)|(df['Month']==11))].index,inplace=True)
```

```python
df.head()
```

```python
date = [str(y)+'-'+str(m)+'-'+str(d) for y, m, d in zip(df.Year, df.Month, df.Day)]
```

```python
df.index = pd.to_datetime(date)
```

```python
df['date'] = df.index
df['dayofyear']=df['date'].dt.dayofyear
df.drop('date',axis=1,inplace=True)
df.head()
```

```python
df.size
```

```python
df["Rainfall Inches"]=df["Rainfall Inches"].astype(float)
```

```python
df.plot(x='Year',y='Rainfall Inches',style='.', figsize=(15,5))
```

```python
train = df.loc[df['Year'] <= 2005]
test = df.loc[df['Year'] == 2020]
```

```python
train.size
```

```python
test.size
```

```python
X_train=train.drop(['Temperature min','Temperature max', 'Humidity 8:30', 'Humidity 2:30', 'Rainfall mm', 'dayofyear'], axis=1)
Y_train=train['Rainfall Inches']
```

```python
X_train.head()
```

```python
X_test=test.drop(['Temperature min','Temperature max', 'Humidity 8:30', 'Humidity 2:30', 'Rainfall mm', 'dayofyear'], axis=1)
Y_test=test['Rainfall Inches']
```

```python
X_test.head(10)
```

```python
X_train.fillna(0, inplace=True)
Y_train.fillna(0, inplace=True)
X_test.fillna(0, inplace=True)
Y_test.fillna(0, inplace=True)
```

```python
from sklearn.svm import SVR
regressor=SVR(gamma='scale',kernel='poly',degree=12)
regressor.fit(X_train,Y_train)
Y_pred = regressor.predict(X_test)
```

```python

df1 = pd.DataFrame({'Actual Rainfall': Y_test, 'Predicted Rainfall': Y_pred})  
df1[df1['Predicted Rainfall']>=1].head(10)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X_test['Day'],Y_test,color='red')
plt.plot(X_test['Day'],Y_test,color='green')
plt.plot(X_test['Day'],Y_pred,color='blue')
```

```python
df1.plot(figsize=(15,4))
df1.plot(subplots=True, figsize=(15,6))
```

```python
from sklearn.metrics import r2_score 
r2 = r2_score(Y_test, Y_pred) 
r2
```
